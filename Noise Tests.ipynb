```python
#locate construction noise
data_descriptor.loc['Noise: Construction Before/After Hours (NM1)'].head(3)
```

```python
#Obtain WHERE most cases of construction noise appear
data_descriptor.loc['Noise: Construction Before/After Hours (NM1)']['Street Name'].value_counts().sort_values().tail(30)
```

```python
#Obtain WHERE most cases of construction noise appear
data_descriptor.loc['Noise: Construction Before/After Hours (NM1)']['Street Name'].value_counts().sort_values().tail(30).plot(kind='bar',alpha=0.75,figsize = (30,10),fontsize=20)
```

```python
#Obtain WHERE most cases of construction noise appear BY ZIP
data_descriptor.loc['Noise: Construction Before/After Hours (NM1)']['Incident Zip'].value_counts().sort_values().tail(30).plot(kind='bar',alpha=0.75,figsize = (30,10),fontsize=20)
```
