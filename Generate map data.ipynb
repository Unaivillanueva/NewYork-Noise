```python
%matplotlib inline
import pandas as pd
```

```python
#load data NYNoise
data = pd.read_csv('NYNoise.csv', index_col=None, encoding='utf-8', dtype={'Incident Zip': str})
data['month'] = pd.DatetimeIndex(data['Created Date']).month
data['year'] = pd.DatetimeIndex(data['Created Date']).year

data = data[data.year == 2016]
```

```python
#create a df with index by ZIP CODE
data_zip = data.groupby(['month', 'Descriptor', 'Incident Zip']).size()
```

```python
from collections import defaultdict
results = defaultdict(lambda: defaultdict(dict))

for index, value in data_zip.iteritems():
    for i, key in enumerate(index):
        if i == 0: nested = results[key]
        elif i == len(index) - 1: nested[key] = value
        else: nested = nested[key]
            
results = {str(z): {str(n): {str(k): int(results[z][n][k]) for k in results[z][n].keys()} for n in results[z].keys()} for z in results.keys()}
```
