```python
%matplotlib inline
import pandas as pd
```

```python
#load data NYNoise
data = pd.DataFrame.from_csv(
    'NYNoise.csv', index_col=None, encoding='utf-8')
```

```python
data_gps = data[data['Incident Address'] != '159-06 UNION TURNPIKE'].loc[:,'Latitude':'Longitude'].dropna(how='any')
```

```python
import geoplotlib
geo_data_for_plotting = {"lat": list(data_gps['Latitude']), "lon": list(data_gps['Longitude'])}
```

```python
data = geoplotlib.utils.DataAccessObject(geo_data_for_plotting)
bbox = geoplotlib.utils.BoundingBox(north=max(geo_data_for_plotting['lat']), west=min(geo_data_for_plotting['lon']), 
                   south=min(geo_data_for_plotting['lat']), east=max(geo_data_for_plotting['lon']))
geoplotlib.set_bbox(bbox)
geoplotlib.kde(data, 5)
geoplotlib.inline()
```

"159-06 UNION TURNPIKE" is clearly an outlier in the data. Lots of alarms noise complaints there - it's a fire department building!!
